```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from matplotlib.colors import LinearSegmentedColormap
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import matplotlib.dates as mdates
from datetime import datetime
```

```python
if 'ts' in df.columns:
    df['timestamp'] = pd.to_datetime(df['ts'], unit='s')
else:
    print("Error: 'ts' column not found in CSV.")
```

```python
def visualize_traffic_flow():
    plt.figure(figsize=(12, 10))

    # make network graphic
    G = nx.DiGraph()

    # add ip address as node
    unique_ips = set(df['Source IP'].unique()) | set(df['Destination IP'].unique())
    for ip in unique_ips:
        G.add_node(ip)

    # count IP set
    ip_pairs = df.groupby(['Source IP', 'Destination IP']).size().reset_index(name='count')

    # add pairs
    max_count = ip_pairs['count'].max()
    for _, row in ip_pairs.iterrows():
        G.add_edge(row['Source IP'], row['Destination IP'], 
                  weight=row['count'], 
                  width=1 + 4 * row['count'] / max_count)

    # set the node location
    pos = nx.spring_layout(G, seed=42)

    # make edge
    edges = G.edges()
    weights = [G[u][v]['width'] for u, v in edges]

    nx.draw_networkx_edges(G, pos, width=weights, alpha=0.7, edge_color='skyblue', 
                           connectionstyle='arc3,rad=0.1', arrowsize=15)

    # draw node
    node_sizes = [1500 if node in df['Source IP'].unique() else 1000 for node in G.nodes()]
    node_colors = ['#ff7f0e' if node in df['Source IP'].unique() else '#1f77b4' for node in G.nodes()]

    nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, alpha=0.8)

    # add title
    plt.title('Network Traffic Flow Between IP Addresses', fontsize=16, fontweight='bold', pad=20)

    # add labels
    if len(G.nodes()) < 50:  # show IP less than 50
        labels = {node: node for node in G.nodes()}
        nx.draw_networkx_labels(G, pos, labels, font_size=10, font_color='black', font_weight='bold', 
                                bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
    else:
        # top 10 IP with high traffic
        top_nodes = df['Source IP'].value_counts().nlargest(10).index.tolist()
        top_nodes += df['Destination IP'].value_counts().nlargest(10).index.tolist()
        labels = {node: node for node in G.nodes() if node in top_nodes}
        nx.draw_networkx_labels(G, pos, labels, font_size=10, font_color='black', font_weight='bold', 
                                bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))



    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', markerfacecolor='#ff7f0e', markersize=10, label='Source IP'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='#1f77b4', markersize=10, label='Destination IP')
    ]
    plt.legend(handles=legend_elements, loc='upper right')

    plt.tight_layout()
    plt.show()  
```

```python
if __name__ == "__main__":
    print("Generating network traffic visualizations...")

    visualize_traffic_flow() 

    print("All visualizations completed!")
```

```python
def visualize_protocol_usage_over_time():
    import matplotlib.dates as mdates

    # Aggregate protocol usage over time (ensure the index is in datetime64 format)
    df['minute'] = pd.to_datetime(df['timestamp']).dt.floor('1min')
    protocol_time_df = df.groupby(['minute', 'Protocol_name']).size().unstack(fill_value=0)

    # convert the index to datetime64 format
    protocol_time_df.index = pd.to_datetime(protocol_time_df.index)

    # Convert datetime to numeric format to ensure compatibility with Matplotlib
    fig, ax = plt.subplots(figsize=(14, 8))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    ax.xaxis.set_major_locator(mdates.AutoDateLocator())

    protocol_time_df.plot.area(ax=ax, alpha=0.7, cmap='viridis')

    plt.title('Protocol Usage Over Time', fontsize=16)
    plt.xlabel('Time', fontsize=12)
    plt.ylabel('Number of Packets', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.legend(title='Protocol', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.show()
```

```python
if __name__ == "__main__":
    print("Generating network traffic visualizations...")

    # Check for any unusual values before converting the timestamp
    print("Before conversion:", df['ts'].describe())

    # Convert the timestamp (force to seconds)
    df['timestamp'] = pd.to_datetime(df['ts'], unit='s')

    # Ensure the timestamp is in datetime64 format after conversion 
    df['timestamp'] = pd.to_datetime(df['timestamp'])

    # Check for any unusual values after conversion
    print("After conversion:", df['timestamp'].describe(datetime_is_numeric=True))

    # Drop rows with NaT values
    df = df.dropna(subset=['timestamp'])

    # Remove abnormal dates (e.g., keep only dates between 2000 and 2030)
    df = df[(df['timestamp'] >= '2000-01-01') & (df['timestamp'] <= '2030-01-01')]

    # Run visualizations with the cleaned data
    visualize_protocol_usage_over_time()  

    print("All visualizations completed!")
```

```python
def visualize_port_scanning():
    import seaborn as sns
    import matplotlib.pyplot as plt

    # Convert if 'timestamp' column is missing
    if 'timestamp' not in df.columns:
        df['timestamp'] = pd.to_datetime(df['ts'], unit='s')

    # prepare the data
    port_scan_df = df[['timestamp', 'Source IP', 'Destination IP', 'Destination Port']].copy()
    port_scan_df['minute'] = port_scan_df['timestamp'].dt.floor('1min')

    #Count the number of unique destination ports per source IP
    port_counts = port_scan_df.groupby(['minute', 'Source IP', 'Destination IP'])['Destination Port'].nunique().reset_index()

    # Check the data
    print("Sample port scan data:\n", port_scan_df.head())
    print("Port count data:\n", port_counts.head())

    # Select the top 20 Source IPs
    interesting_ips = port_counts.groupby('Source IP')['Destination Port'].sum().nlargest(20).index
    port_scan_pivot = port_counts[port_counts['Source IP'].isin(interesting_ips)]

    if port_scan_pivot.empty:
        print(" Not enough data for port scanning visualization, but showing empty heatmap.")

    #  plot heatmap
    pivot_table = port_scan_pivot.pivot_table(
        index='Source IP', 
        columns='minute', 
        values='Destination Port',
        aggfunc='sum',
        fill_value=0
    )

    plt.figure(figsize=(16, 8))
    sns.heatmap(pivot_table, cmap='YlOrRd', annot=False, cbar_kws={'label': 'Unique Ports Accessed'})

    plt.title('Potential Port Scanning Activity Over Time', fontsize=16)
    plt.xlabel('Time', fontsize=12)
    plt.ylabel('Source IP Address', fontsize=12)
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()  

if __name__ == "__main__":
    print("Generating network traffic visualizations...")

    visualize_port_scanning()

    print("All visualizations completed!")
```

```python
def visualize_tcp_flag_patterns():
    import numpy as np
    import matplotlib.pyplot as plt

    # Check TCP flag related columns
    tcp_flags = ['fin_flag_number', 'syn_flag_number', 'rst_flag_number', 
                'psh_flag_number', 'ack_flag_number', 'urg_flag_number', 
                'ece_flag_number', 'cwr_flag_number']

    # Check if the TCP flag columns exist in the data
    missing_flags = [flag for flag in tcp_flags if flag not in df.columns]
    if missing_flags:
        print(f"Missing TCP flag columns: {missing_flags}")
        return 

    # Calculate the frequency of flag usage
    flag_counts = df[tcp_flags].sum()

    # check the data
    print("TCP flag usage sum:\n", flag_counts)

    # Ensure graph is plotted even if all TCP flag values are 0
    if flag_counts.sum() == 0:
        print("No significant TCP flag data available, but generating an empty radar chart.")
        flag_counts = pd.Series([1] * len(tcp_flags), index=tcp_flags)  

    # Prepare radar chart data
    categories = [flag.replace('_flag_number', '').upper() for flag in tcp_flags]
    values = flag_counts.values.tolist()
    values.append(values[0]) 
    categories.append(categories[0])  

    # calculate angles
    N = len(categories) - 1
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += [angles[0]]  

    # make a graph
    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

    # fill the polygon
    ax.fill(angles, values, color='skyblue', alpha=0.25)

    # add line
    ax.plot(angles, values, color='blue', linewidth=2)

    # add label
    ax.set_thetagrids(np.degrees(angles), categories)

    # add title
    ax.set_title('TCP Flag Usage Patterns', fontsize=16)

    plt.tight_layout()
    plt.show() 

if __name__ == "__main__":
    print("Generating network traffic visualizations...")

    visualize_tcp_flag_patterns()

    print("All visualizations completed!")
```

```python
def visualize_packet_size_distribution():
    import seaborn as sns
    import matplotlib.pyplot as plt

    # Check for 'Protocol_name' and 'Tot size' columns
    if 'Protocol_name' not in df.columns or 'Tot size' not in df.columns:
        print(" Missing required columns: 'Protocol_name' or 'Tot size'")
        return

    # Remove NaN values
    df_cleaned = df.dropna(subset=['Protocol_name', 'Tot size'])

    # check the data
    print("Packet size stats:\n", df_cleaned['Tot size'].describe())

    # Handle the case when there's no data
    if df_cleaned.empty:
        print("No data available for packet size distribution.")
        return

    # Visualization
    plt.figure(figsize=(12, 8))

    # Packet size distribution by protocol
    protocols = df_cleaned['Protocol_name'].unique()
    
    for protocol in protocols:
        subset = df_cleaned[df_cleaned['Protocol_name'] == protocol]
        if not subset.empty:
            sns.histplot(subset['Tot size'], bins=30, alpha=0.5, label=protocol, kde=True)

    plt.title('Packet Size Distribution by Protocol', fontsize=16)
    plt.xscale('log')  

    plt.xlabel('Packet Size (bytes)', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.legend(title='Protocol')

    plt.tight_layout()
    plt.show() 

if __name__ == "__main__":
    print("Generating network traffic visualizations...")

    visualize_packet_size_distribution()

    print("All visualizations completed!")
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

def visualize_geographic_traffic(sample=True):
    if sample:
        # Generate sample data (For real implementation, IP geolocation API is required)
        locations = pd.DataFrame({
            'ip': df['Source IP'].unique(),
            'lat': np.random.uniform(20, 60, len(df['Source IP'].unique())),
            'lon': np.random.uniform(-120, 120, len(df['Source IP'].unique())),
            'country': np.random.choice(['USA', 'China', 'Russia', 'UK', 'Germany', 'Brazil', 'India', 'Japan'], 
                                       len(df['Source IP'].unique()))
        })
        
        # Generate connection data
        connections = df.groupby(['Source IP', 'Destination IP']).size().reset_index(name='count')
        connections = connections.merge(
            locations, left_on='Source IP', right_on='ip'
        ).rename(columns={'lat': 'src_lat', 'lon': 'src_lon', 'country': 'src_country'})

        # Generate location data for destination IPs
        dest_locations = pd.DataFrame({
            'ip': df['Destination IP'].unique(),
            'lat': np.random.uniform(20, 60, len(df['Destination IP'].unique())),
            'lon': np.random.uniform(-120, 120, len(df['Destination IP'].unique())),
            'country': np.random.choice(['USA', 'China', 'Russia', 'UK', 'Germany', 'Brazil', 'India', 'Japan'], 
                                        len(df['Destination IP'].unique()))
        })
        
        connections = connections.merge(
            dest_locations, left_on='Destination IP', right_on='ip'
        ).rename(columns={'lat': 'dst_lat', 'lon': 'dst_lon', 'country': 'dst_country'})

        # Output sample data for verification
        print("Generated IP locations sample data:\n", locations.head())
        print("Generated connections sample data:\n", connections.head())

        # Plotly geographic visualization
        fig = px.scatter_geo(
            locations, 
            lat='lat', 
            lon='lon',
            color='country',
            hover_name='ip',
            size=np.random.randint(10, 50, len(locations)),  
            projection='natural earth'
        )

        # Add connections (network traffic flow lines)
        for _, row in connections.iterrows():
            fig.add_trace(
                go.Scattergeo(
                    lon=[row['src_lon'], row['dst_lon']],
                    lat=[row['src_lat'], row['dst_lat']],
                    mode='lines',
                    line=dict(width=1, color='red'),
                    opacity=0.5,
                    showlegend=False
                )
            )

        fig.update_layout(
            title='Geographic Distribution of Network Traffic (Simulated)',
            geo=dict(
                showland=True,
                landcolor='rgb(243, 243, 243)',
                countrycolor='rgb(204, 204, 204)',
                showocean=True,
                oceancolor='rgb(230, 230, 250)'
            )
        )

        # Save as HTML file and display the output location
        output_file = "geographic_traffic_map.html"
        fig.write_html(output_file)
        print(f"Map saved as {output_file}")

    else:
        print("Geographic visualization requires IP geolocation which is not included in this script")

if __name__ == "__main__":
    print("Generating network traffic visualizations...")

    visualize_geographic_traffic(sample=True)

    print("All visualizations completed!")
```

```python
def visualize_correlation_matrix():
    #Select numeric columns that are useful for correlation analysis
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
    
    #Exclude columns such as IP addresses, timestamps, labels, etc.
    exclude_cols = ['ts', 'flow_duration', 'Source IP', 'Destination IP', 'Source Port', 
                    'Destination Port', 'Protocol Type', 'Protocol_name', 'label', 'subLabel', 'subLabelCat']
    
    # Filter out excluded columns from the correlation analysis
    corr_cols = [col for col in numeric_cols if col not in exclude_cols]
    
    # If there are too many columns, select only the first 15
    if len(corr_cols) > 15:
        corr_cols = corr_cols[:15]
    
    # Compute the correlation matrix
    corr_matrix = df[corr_cols].corr()
    
    # Plot the heatmap
    plt.figure(figsize=(14, 12))
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  
    
    # 사Create a mask to hide the upper triangle of the matrix 
    cmap = LinearSegmentedColormap.from_list("correlation_cmap", ["#4169E1", "white", "#FF4500"])
    
    sns.heatmap(corr_matrix, mask=mask, cmap=cmap, vmin=-1, vmax=1, center=0,
                annot=True, fmt=".2f", square=True, linewidths=.5, cbar_kws={"shrink": .8})
    
    plt.title('Feature Correlation Matrix', fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    
    plt.tight_layout()
    plt.savefig('correlation_matrix.png', dpi=300)
    plt.close()
```

```python
if __name__ == "__main__":
    print("Generating network traffic visualizations...")

    visualize_correlation_matrix()

    print("All visualizations completed!")
```

```python
# Create timestamp column
df['timestamp'] = pd.to_datetime(df['ts'], unit='s')

# Group by 1-minute intervals and count the requests
df['minute'] = df['timestamp'].dt.floor('1min')
request_per_min = df.groupby('minute').size().reset_index(name='requests')

# Set the anomaly detection threshold
threshold = request_per_min['requests'].mean() + 2 * request_per_min['requests'].std()
suspicious_df = request_per_min[request_per_min['requests'] > threshold]

# visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
plt.plot(request_per_min['minute'], request_per_min['requests'], label='Requests/min', color='skyblue')
plt.scatter(suspicious_df['minute'], suspicious_df['requests'], color='red', label='Anomalous Requests')
plt.axhline(threshold, color='orange', linestyle='--', label='Threshold (Mean + 2σ)')
plt.title("Requests Over Time with DDoS Detection", fontsize=16)
plt.xlabel("Time")
plt.ylabel("Requests per Minute")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("ddos_detection_static.png", dpi=300)
plt.show()
```
